## 第一个Python程序

```python
# 通常来讲，第一个Python程序是打印Hello World
print("Hello World")
```

print()是一个函数，用来将括号中的内容打印出来。而Hello World是一段字符串，在Python语言中，可以使用""或者''来表示字符串，但是不能够使用"'或者'"来表示字符串，另外需要注意符号是英文的符号，不要使用其他语言的符号。

```python
# 输出重复的内容
print("*" * 10)  # 可以将 * 符号重复十次
```

格式化输出是一种继承自C语言的输出方式，使用f"{}"的形式来进行，括号中是需要被格式化的内容。

```python
pi = 3.1415926
```

```python
# 保留几位数
print(f"π的值是{pi:.3}")  # 保留三位数
```

```python
# 保留小数点后几位数
print(f"π的值是{pi:.3f}")  # 添加了f，从而指明这是小数点后保留的位数
```

通过this模块，可以打印出来Python语言的禅：

```python
import this
print(this)
```
