## Stacks

Stack的定义为：Last in - First out。

如下图：





```python
# 使用list来模拟
stack_list = [1, 2, 3, 4, 5]

# last in
stack_list.append(6)
```

```python
stack_list
```

```python
# first out
stack_list.pop()
```

使用Python来实现Stack：

```python
# 节点生成
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
class Stack:
    
    # 实例化一个对象
    def __init__(self, value):
        new_node = Node(value)
        self.top = new_node  # 最顶层
        self.height = 1  # 大小
    
    # 打印内容
    def print_stack(self):
        temp = self.top
        while temp is not None:
            print(temp.value)
            temp = temp.next
    
    # 推入
    def push(self, value):
        new_node = Node(value)  # 实例化一个Node
        if self.height == 0:  # 考虑为空的情况
            self.top == new_node 
        else:
            new_node.next = self.top
            self.top = new_node  # 让top节点为新生成的节点
        self.height += 1  # 修改长度、
    
    # 弹出
    def pop(self):
        
        # 如果为空，则返回None
        if self.height == 0:
            return None
        
        # 如果不为空
        temp = self.top  # 临时变量为stack的top节点
        self.top = self.top.next  # 将当前top修改为当前top的next的节点
        temp.next = None  # 让临时变量的next节点为None
        
        self.height -= 1  # 修改stack的长度
        
        return temp  # 返回temp
```

```python
# testing
my_stack = Stack(1)

my_stack.push(2)
my_stack.push(3)
my_stack.push(4)
```

```python
my_stack.print_stack()
```

```python
pop_ele = my_stack.pop()
```

```python
my_stack.print_stack()
```

```python
pop_ele.value
```

## Queues

队列的逻辑是：First in - First out。

如下图所示：



```python
from collections import deque

queue = deque([])

queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
queue.append(5)

print(queue)
```

```python
# 弹出
pop_ele_l = queue.popleft()
pop_ele_l
```

```python
# 测试一下 right
pop_ele_r = queue.popright()
```

可以发现不能够将右侧的元素弹出。

使用Python来实现Queue：

```python
# 同样还是创建Node类

class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
class Queue:
    def __init__(self, value):
        new_node = Node(value)
        self.first = new_node
        self.last = new_node
        self.length = 1
    
    def print_queue(self):
        temp = self.first
        while temp is not None:
            print(temp.value)
            temp = temp.next
            
    def enqueue(self, value):
        # 调用Node类来生成一个新的结点
        new_node = Node(value)
        # 考虑队列为空的情况
        if self.first is None:
            self.first = new_node
            self.lasst = new_node
        else:
            self.last.next = new_node  # 队列的last结点指向新结点
            self.last = new_node  # 让新结点成为队列的新last结点
        self.length += 1  # 修改队列的大小
        
    def dequeue(self):
        # 考虑到队列的长度为0的情况
        if self.length == 0:
            return None
        # 一个临时变量
        temp = self.first
        # 只有一个结点的队列
        if self.length == 1:
            self.first = None
            self.last = None
        # 队列不止一个结点的情况
        else:
            self.first = self.first.next  # 让队列的first结点变成队列的第二个结点
            temp.next = None  # 然temp指向None
        self.length -= 1  # 修改队列的长度
        return temp  # 返回被弹出的结点
```

```python
my_queue = Queue(1)
my_queue.enqueue(2)
```

```python
my_queue.print_queue()
```

```python
pop_ele = my_queue.dequeue()
```

```python
pop_ele.value
```

```python
my_queue.print_queue()
```

## Sets

集合是无序且不重复元素的集合，用{}来进行表示。

三种括号：
- [] : 表示 list
- () : 表示 tuple
- {} ：表示set

```python
numbers = [1, 1, 2, 2, 3, 4, 5]

# 转换为set
numbers_set = set(numbers)
# 打印
print(numbers_set)
```

可以发现重复的元素被自动的删除了。

集合的方法如下所示：
- .add()：添加元素，如果已经存在，则会自动删除
- .update()：添加元素，元素的类型不限
- .remove()：移除元素，如果元素不存在，则会报错
- .discard()：移除元素，如果元素不存在，也不会报错
- .pop()：弹出元素，随机
- .clear()：清楚集合中的所有元素

集合提供了一些对两个集合进行操作的方法：

```python
set_01 = {1, 2, 3, 4, 5, 6, 7}
set_02 = {5, 6, 7, 8, 9, 0}
```

```python
print(set_01 | set_02)  # 并集
```

```python
print(set_01 & set_02)  # 交集
```

```python
print(set_01 - set_02)  # 在set_01中，但是不在set_02中
```

```python
print(set_01 ^ set_02)  # 交集的反集
```
