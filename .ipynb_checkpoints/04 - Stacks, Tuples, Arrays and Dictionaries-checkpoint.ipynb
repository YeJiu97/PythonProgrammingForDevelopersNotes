PythonStack

```python
# 
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
class Stack:
    
    # 
    def __init__(self, value):
        new_node = Node(value)
        self.top = new_node  # 
        self.height = 1  # 
    
    # 
    def print_stack(self):
        temp = self.top
        while temp is not None:
            print(temp.value)
            temp = temp.next
    
    # 
    def push(self, value):
        new_node = Node(value)  # Node
        if self.height == 0:  # 
            self.top == new_node 
        else:
            new_node.next = self.top
            self.top = new_node  # top
        self.height += 1  # 
    
    # 
    def pop(self):
        
        # None
        if self.height == 0:
            return None
        
        # 
        temp = self.top  # stacktop
        self.top = self.top.next  # toptopnext
        temp.next = None  # nextNone
        
        self.height -= 1  # stack
        
        return temp  # temp
```

```python
# testing
my_stack = Stack(1)

my_stack.push(2)
my_stack.push(3)
my_stack.push(4)
```

```python
my_stack.print_stack()
```

```python
pop_ele = my_stack.pop()
```

```python
my_stack.print_stack()
```

```python
pop_ele.value
```

## Queues

First in - First out





```python
from collections import deque

queue = deque([])

queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
queue.append(5)

print(queue)
```

```python
# 
pop_ele_l = queue.popleft()
pop_ele_l
```

```python
#  right
pop_ele_r = queue.popright()
```

PythonQueue

```python
# Node

class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
class Queue:
    def __init__(self, value):
        new_node = Node(value)
        self.first = new_node
        self.last = new_node
        self.length = 1
    
    def print_queue(self):
        temp = self.first
        while temp is not None:
            print(temp.value)
            temp = temp.next
            
    def enqueue(self, value):
        # Node
        new_node = Node(value)
        # 
        if self.first is None:
            self.first = new_node
            self.lasst = new_node
        else:
            self.last.next = new_node  # last
            self.last = new_node  # last
        self.length += 1  # 
        
    def dequeue(self):
        # 0
        if self.length == 0:
            return None
        # 
        temp = self.first
        # 
        if self.length == 1:
            self.first = None
            self.last = None
        # 
        else:
            self.first = self.first.next  # first
            temp.next = None  # tempNone
        self.length -= 1  # 
        return temp  # 
```

```python
my_queue = Queue(1)
my_queue.enqueue(2)
```

```python
my_queue.print_queue()
```

```python
pop_ele = my_queue.dequeue()
```

```python
pop_ele.value
```

```python
my_queue.print_queue()
```
